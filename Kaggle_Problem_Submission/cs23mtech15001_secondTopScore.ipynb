```python
import pandas as pd
import os
import numpy as np
from sklearn import preprocessing

from scipy import stats as st
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

from sklearn.ensemble import RandomForestClassifier

from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("./data/iith_foml_2023_train.csv")
```

```python
class_label_freq_dict = data['Target Variable (Discrete)'].value_counts().to_dict()
total_samples = data.shape[0]
class_weights_dict = {}
for class_label_name,class_freq in class_label_freq_dict.items():
    class_weights_dict[class_label_name] = total_samples/class_label_freq_dict[class_label_name]


print(class_weights_dict)
```

```python
# all Discrete variables does not have any NaN values 
data.isnull().any()
```

```python
# filling the NaN values with mean values for the continous variables which has NaN values
data = data.fillna(data.mean())
```

```python
for column_name in data.columns.tolist():
    if 'Discrete' in column_name:
        print("For column name ",column_name)
        print("Number of unique values ",len(data[column_name].unique().tolist()))
        # print(data[column_name].value_counts())
        print()
```

```python
total_columns = data.columns.tolist()
continous_feature_columns = []

print(total_columns)
for column_name in total_columns:
    if 'Discrete' not in column_name:
        continous_feature_columns.append(column_name)

print()
print(continous_feature_columns)
```

```python
# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Scale the selected columns
scaled_data = pd.DataFrame(scaler.fit_transform(data[continous_feature_columns]), columns=continous_feature_columns)

# Concatenate the scaled columns with the rest of the DataFrame
data_preprocessed = pd.concat([data.drop(continous_feature_columns, axis=1), scaled_data], axis=1)


data_preprocessed.head()
```

```python
data_preprocessed.columns
```

# Training the Model

```python

X_train,y_train = data.iloc[:,:-1],data.iloc[:,-1]

rf = RandomForestClassifier(random_state=42,class_weight=class_weights_dict)
rf.fit(X_train, y_train)
```

# Predicting on Test data

```python
data_test = pd.read_csv("./data/test_input.csv")

data_test.head()
```

```python
# all Discrete variables does not have any NaN values 
data_test.isnull().any()
```

```python
data_test = data_test.fillna(data.mean())
```

```python
data_submission_final = pd.DataFrame()

index_list = data_test.index.tolist()

index_list = [x + 1 for x in index_list]
data_submission_final['Id'] = index_list
data_submission_final['Category'] = rf.predict(data_test)
```

```python
data_submission_final.to_csv("test_output_secondTopScore.csv",index=False)
```
